```python
import pandas as pd

df = pd.read_csv('../../data/raw/fake_job_postings.csv')
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['text'] = (
    df['title'].fillna('') + ' ' +
    df['description'].fillna('') + ' ' +
    df['requirements'].fillna('')
)
```

```python
import re

df['has_salary'] = df['salary_range'].notna().astype(int)

df['urgency_words'] = df['text'].str.count(
    r'\b(urgent|immediately|asap|earn \$)\b', flags=re.I
)

df['generic_email'] = df['text'].str.contains(
    r'@gmail|@yahoo|@hotmail', flags=re.I
).astype(int)

df['all_caps_ratio'] = df['text'].apply(
    lambda x: sum(1 for c in x if c.isupper()) / max(len(x), 1)
)
```

```python
df['text_length'] = df['text'].apply(len)
```

```python
df[['text', 'has_salary', 'urgency_words', 'generic_email', 'all_caps_ratio']].head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(
    df,
    test_size=0.2,
    stratify=df['fraudulent'],
    random_state=42
)

train, val = train_test_split(
    train,
    test_size=0.15,
    stratify=train['fraudulent'],
    random_state=42
)
```

```python
train.to_csv('../../data/processed/train.csv', index=False)
val.to_csv('../../data/processed/val.csv', index=False)
test.to_csv('../../data/processed/test.csv', index=False)
```
